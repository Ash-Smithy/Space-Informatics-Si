```python
from newspaper import Article
import random
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')
```

```python
nltk.download('punkt', quiet=True)
```

```python
article = Article('https://en.wikipedia.org/wiki/Organic_food')
article.download()
article.parse()
article.nlp()
corpus = article.text 
```

```python
#cheking corpus
print(corpus)
```

```python
#Tokenization
text = corpus
sentence_list = nltk.sent_tokenize(text) #Gives list of sentences
```

```python
#Print sentences
print(sentence_list)
```

```python
#Function to greet user
def greet(text):
    text = text.lower()

    bot_greets = ['Hi!','Hello!','Greetings!','Good Day!','Hello there!']
    user_greets = ['hi','hey','hello','greetings','sup','wassup']
    user_ques = ['How are you?']
    bot_rep = ['I am good, hope it\'s the same with you!','All good, thank you.']
    for word in text.split():
        if word in user_greets:
            return random.choice(bot_greets)
        elif word in user_ques:
            return random.choice(bot_rep)
```

```python
#Function to sort
def index_sort(list_var):
    length = len(list_var)
    list_index = list(range(0,length))
    
    x = list_var
    for i in range(length):
        for j in range(length):
            if x[list_index[i]] > x[list_index[j]]:
                #Swap
                temp = list_index[i]
                list_index[i] = list_index[j]
                list_index[j] = temp
    return list_index
```

```python
#Bot response
def bot_response(user_input):
    user_input = user_input.lower()
    sentence_list.append(user_input)
    bot_response = ''
    cm = CountVectorizer().fit_transform(sentence_list) #count matrix
    similarty_scores = cosine_similarity(cm[-1], cm) #compare last sentence of user input with everything in cm
    similarty_scores_list = similarty_scores.flatten()
    index = index_sort(similarty_scores_list)
    index = index[1:]
    flag = 0

    j = 0
    for i in range(len(index)):
        if similarty_scores_list[index[i]] > 0.0:
            bot_response = bot_response+' '+sentence_list[index[i]]
            flag = 1
            j = j+1
        if j > 2:
            break;
    if flag == 0:
        bot_response = bot_response+' '+'Sorry I did not understand.'

    sentence_list.remove(user_input)
    return bot_response
```

```python
#chat 
print('Bot: Hi! How can I help you? (To exit, type "bye")')

exit_list = ['exit','bye','quit','end','break']
while(True):
    user_input = input()
    if user_input.lower() in exit_list:
        print('Bot: See you later ~ alligator!')
        break;
    else:
        if greet(user_input) != None:
            print('Bot: '+greet(user_input))
        else:
            print('Bot: '+bot_response(user_input))
```
